–í–∏–∫–æ–Ω—É—î—Ç—å—Å—è –Ω–∞—Ç—É—Ä–Ω–∏–π –µ–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç, –ø—ñ–¥ —á–∞—Å —è–∫–æ–≥–æ **10 –∞—Ä—Ç–∏–ª–µ—Ä—ñ–π—Å—å–∫–∏—Ö –±–∞—Ç–∞—Ä–µ–π** –≤–∏–∫–æ–Ω—É—é—Ç—å –∑–∞–≤–¥–∞–Ω–Ω—è –∑ –ø—ñ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–æ –±–æ–π–æ–≤–∏—Ö –¥—ñ–π. –ß–∞—Å –ø—ñ–¥–≥–æ—Ç–æ–≤–∫–∏ –∫–æ–∂–Ω–æ—ó –±–∞—Ç–∞—Ä–µ—ó –∑–∞–ø–∏—Å–∞–Ω–æ (–≤ —Ö–≤–∏–ª–∏–Ω–∞—Ö).  

–î–∞–Ω—ñ:  
\[
T = [32, 39, 47, 52, 38, 54, 28, 43, 69, 48]
\]

**–ó–∞–≤–¥–∞–Ω–Ω—è:**  
1. –í–∏–∑–Ω–∞—á–∏—Ç–∏ —Å–µ—Ä–µ–¥–Ω—ñ–π —á–∞—Å –ø—ñ–¥–≥–æ—Ç–æ–≤–∫–∏ –±–∞—Ç–∞—Ä–µ—ó.  
2. –í–∏–∑–Ω–∞—á–∏—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É –ø–æ—Ö–∏–±–∫—É –ø—Ä–∏ –ø—Ä–∏–π–Ω—è—Ç—Ç—ñ —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ —á–∞—Å—É –∑–∞ –º–∞—Ç–µ–º–∞—Ç–∏—á–Ω–µ —Å–ø–æ–¥—ñ–≤–∞–Ω–Ω—è.  
3. –í–∏–∑–Ω–∞—á–∏—Ç–∏ –Ω–µ–æ–±—Ö—ñ–¥–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å –±–∞—Ç–∞—Ä–µ–π –¥–ª—è —Ç–æ—á–Ω–æ–≥–æ –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ —á–∞—Å—É.  

**–†–µ–∞–ª—ñ–∑–∞—Ü—ñ—è –≤ Python:**  
```python
import numpy as np

# –í–∏—Ö—ñ–¥–Ω—ñ –¥–∞–Ω—ñ
T = np.array([32, 39, 47, 52, 38, 54, 28, 43, 69, 48])

# –û–±—á–∏—Å–ª–µ–Ω–Ω—è —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ —á–∞—Å—É –ø—ñ–¥–≥–æ—Ç–æ–≤–∫–∏
mean_T = np.mean(T)

print(f"–°–µ—Ä–µ–¥–Ω—ñ–π —á–∞—Å –ø—ñ–¥–≥–æ—Ç–æ–≤–∫–∏ –±–∞—Ç–∞—Ä–µ—ó: {mean_T:.2f} —Ö–≤.")
```
‚úÖ **–†–µ–∑—É–ª—å—Ç–∞—Ç:** 45 —Ö–≤.

### **2.2. –†—ñ—à–µ–Ω–Ω—è –∑–∞–¥–∞—á—ñ (30 —Ö–≤.)**

#### **2.2.1. –û–±—á–∏—Å–ª–µ–Ω–Ω—è –¥–∏—Å–ø–µ—Ä—Å—ñ—ó —Ç–∞ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è**
–§–æ—Ä–º—É–ª–∏:  
- **–î–∏—Å–ø–µ—Ä—Å—ñ—è**:
  \[
  D(T) = \frac{1}{N-1} \sum (T_i - \bar{T})^2
  \]
- **–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–µ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è**:
  \[
  \sigma = \sqrt{D(T)}
  \]

```python
# –û–±—á–∏—Å–ª–µ–Ω–Ω—è –¥–∏—Å–ø–µ—Ä—Å—ñ—ó —Ç–∞ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è
variance_T = np.var(T, ddof=1)  # –î–∏—Å–ø–µ—Ä—Å—ñ—è
std_T = np.sqrt(variance_T)  # –°–ö–í

print(f"–î–∏—Å–ø–µ—Ä—Å—ñ—è: {variance_T:.2f} —Ö–≤¬≤")
print(f"–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–µ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è: {std_T:.2f} —Ö–≤.")
```
‚úÖ **–†–µ–∑—É–ª—å—Ç–∞—Ç:**  
- –î–∏—Å–ø–µ—Ä—Å—ñ—è: 140.67 —Ö–≤¬≤  
- –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–µ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è: 11.86 —Ö–≤.  

---

#### **2.2.2. –û–±—á–∏—Å–ª–µ–Ω–Ω—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ—ó –ø–æ—Ö–∏–±–∫–∏**  
–§–æ—Ä–º—É–ª–∞ –ø–æ—Ö–∏–±–∫–∏:

\[
\epsilon = \frac{k \cdot \sigma}{\sqrt{N}}
\]

–¥–µ **\( k = 2 \)** –¥–ª—è –¥–æ–≤—ñ—Ä—á–æ—ó –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç—ñ **0.955**.  

```python
# –ü–∞—Ä–∞–º–µ—Ç—Ä–∏
k = 2
N = len(T)

# –û–±—á–∏—Å–ª–µ–Ω–Ω—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ—ó –ø–æ—Ö–∏–±–∫–∏
error_max = k * std_T / np.sqrt(N)

print(f"–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞ –ø–æ—Ö–∏–±–∫–∞: {error_max:.2f} —Ö–≤.")
```
‚úÖ **–†–µ–∑—É–ª—å—Ç–∞—Ç:** 7.50 —Ö–≤.

---

### **2.3. –í–∏–∑–Ω–∞—á–µ–Ω–Ω—è –Ω–µ–æ–±—Ö—ñ–¥–Ω–æ—ó –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –±–∞—Ç–∞—Ä–µ–π (30 —Ö–≤.)**
–©–æ–± –∑–º–µ–Ω—à–∏—Ç–∏ –ø–æ—Ö–∏–±–∫—É –¥–æ **5 —Ö–≤.**, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î–º–æ —Ñ–æ—Ä–º—É–ª—É:

\[
N = \left( \frac{k \cdot \sigma}{\epsilon_{\text{–∑–∞–¥}}} \right)^2
\]

```python
# –ó–∞–¥–∞–Ω–∞ –ø–æ—Ö–∏–±–∫–∞
epsilon_target = 5

# –û–±—á–∏—Å–ª–µ–Ω–Ω—è –Ω–µ–æ–±—Ö—ñ–¥–Ω–æ—ó –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –±–∞—Ç–∞—Ä–µ–π
N_required = (k * std_T / epsilon_target) ** 2
N_required = np.ceil(N_required)  # –û–∫—Ä—É–≥–ª–µ–Ω–Ω—è –¥–æ —Ü—ñ–ª–æ–≥–æ —á–∏—Å–ª–∞

print(f"–ù–µ–æ–±—Ö—ñ–¥–Ω–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –±–∞—Ç–∞—Ä–µ–π: {N_required:.0f}")
```
‚úÖ **–†–µ–∑—É–ª—å—Ç–∞—Ç:** 23 –±–∞—Ç–∞—Ä–µ—ó.

## **1. –í–∏–∑–Ω–∞—á–µ–Ω–Ω—è —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ —á–∞—Å—É –ø—ñ–¥–≥–æ—Ç–æ–≤–∫–∏ –±–∞—Ç–∞—Ä–µ—ó**  

### **–§–æ—Ä–º—É–ª–∞ —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è (–º–∞—Ç–µ–º–∞—Ç–∏—á–Ω–µ —Å–ø–æ–¥—ñ–≤–∞–Ω–Ω—è)**
\[
\bar{T} = \frac{1}{N} \sum_{i=1}^{N} T_i
\]
–¥–µ:  
- \( \bar{T} \) ‚Äì —Å–µ—Ä–µ–¥–Ω—ñ–π —á–∞—Å –ø—ñ–¥–≥–æ—Ç–æ–≤–∫–∏,  
- \( N \) ‚Äì –∫—ñ–ª—å–∫—ñ—Å—Ç—å –±–∞—Ç–∞—Ä–µ–π,  
- \( T_i \) ‚Äì —á–∞—Å –ø—ñ–¥–≥–æ—Ç–æ–≤–∫–∏ \( i \)-—Ç–æ—ó –±–∞—Ç–∞—Ä–µ—ó.  

### **–ü–æ—è—Å–Ω–µ–Ω–Ω—è:**  
–¶—è —Ñ–æ—Ä–º—É–ª–∞ –æ–±—á–∏—Å–ª—é—î **—Å–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è** —á–∞—Å—É –ø—ñ–¥–≥–æ—Ç–æ–≤–∫–∏. –ú–∏ –ø—Ä–æ—Å—Ç–æ –ø—ñ–¥—Å—É–º–æ–≤—É—î–º–æ –≤—Å—ñ –∑–Ω–∞—á–µ–Ω–Ω—è —á–∞—Å—É –ø—ñ–¥–≥–æ—Ç–æ–≤–∫–∏ —ñ –¥—ñ–ª–∏–º–æ —ó—Ö –Ω–∞ –∑–∞–≥–∞–ª—å–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å –±–∞—Ç–∞—Ä–µ–π.

#### **–ü—Ä–∏–∫–ª–∞–¥ —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É:**
–ù–µ—Ö–∞–π —î —Ç–∞–∫—ñ –¥–∞–Ω—ñ –ø—Ä–æ –ø—ñ–¥–≥–æ—Ç–æ–≤–∫—É –±–∞—Ç–∞—Ä–µ–π:
\[
T = [32, 39, 47, 52, 38, 54, 28, 43, 69, 48]
\]

–¢–æ–¥—ñ —Å–µ—Ä–µ–¥–Ω—ñ–π —á–∞—Å:
\[
\bar{T} = \frac{32 + 39 + 47 + 52 + 38 + 54 + 28 + 43 + 69 + 48}{10} = \frac{450}{10} = 45 \text{ —Ö–≤.}
\]

## **2. –í–∏–∑–Ω–∞—á–µ–Ω–Ω—è –¥–∏—Å–ø–µ—Ä—Å—ñ—ó —Ç–∞ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è**  

### **–§–æ—Ä–º—É–ª–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—ó**  
\[
D(T) = \frac{1}{N - 1} \sum_{i=1}^{N} (T_i - \bar{T})^2
\]
–¥–µ:  
- \( D(T) \) ‚Äì –¥–∏—Å–ø–µ—Ä—Å—ñ—è,  
- \( (T_i - \bar{T})^2 \) ‚Äì –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è –∫–æ–∂–Ω–æ–≥–æ —á–∞—Å—É –ø—ñ–¥–≥–æ—Ç–æ–≤–∫–∏ –≤—ñ–¥ —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ —É –∫–≤–∞–¥—Ä–∞—Ç—ñ,
